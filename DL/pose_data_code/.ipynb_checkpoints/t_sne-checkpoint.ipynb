```python
import numpy as np
import pandas as pd
import os

csv_dir = r'C:\Users\comoj\SMART_Referee\data\output.csv'
col_names = ['class', 'nose_x', 'nose_y', 'neck_x', 'neck_y', 'rshoulder_x', 'rshoulder_y', 'relbow_x', 'relbow_y',
             'rwrist_x', 'rwrist_y', 'lshoulder_x', 'lshoulder_y', 'lelbow_x', 'lelbow_y', 'lwrist_x', 'lwrist_y',
             'midhip_x', 'midhip_y', 'rhip_x', 'rhip_y', 'rknee_x', 'rknee_y', 'rankle_x', 'rankle_y',
             'lhip_x', 'lhip_y', 'lknee_x', 'lknee_y', 'lankle_x', 'lankle_y', 'reye_x', 'reye_y',
             'leye_x', 'leye_y', 'rear_x', 'rear_y', 'lear_x', 'lear_y', 'lbigtoe_x', 'lbigtoe_y',
             'lsmalltoe_x', 'lsmalltoe_y', 'lheel_x', 'lhell_y', 'rbigtoe_x', 'rbigtoe_y', 'rsmalltoe_x', 'rsmalltoe_y',
             'rheel_x', 'rheel_y']
```

```python
df = pd.read_csv(csv_dir, names=col_names)

new_df = pd.DataFrame( columns=col_names)
new_df['class'] = df['class']

def cal_euclidean(col3, col4, col1=df['neck_x'], col2=df['neck_y']):
    normal = np.sqrt((col1 - col3)**2 + (col2 - col4)**2)
    col3 = (col1-col3)/normal
    col4 = (col2-col4)/normal
    return col3, col4
df.head()
```

```python
df['shoulder_dist'] = np.sqrt((df['rshoulder_x'] - df['lshoulder_x'])**2 + (df['rshoulder_y'] - df['lshoulder_y'])**2)/25
df[['rshoulder_x', 'rshoulder_y', 'lshoulder_x', 'lshoulder_y', 'shoulder_dist']].head()
```

```python
sample1_df = df[['class', 'shoulder_dist']]
sample1_df.rename(columns = {'class' : 'c'}, inplace = True)
sample1_df = sample1_df.dropna(axis = 0)

from statsmodels.stats.weightstats import ztest, ttest_ind

result = []
for i in range(10, len(sample1_df)):
    v = ttest_ind(sample_df['shoulder_dist'],sample1_df.iloc[i-10:i]['shoulder_dist'])[1]#, usevar='unequal')[1]
    print(ttest_ind(sample_df['shoulder_dist'],sample1_df.iloc[i-10:i]['shoulder_dist']))#, usevar='unequal'))
    print(sample1_df.iloc[i].name, 
          v, v > 0.05,
          sample1_df.iloc[i]['c'])
    if v>0.05:
        result.append(sample1_df.iloc[i].name)

print(result)
    
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
pca.fit(pca_df)
print('singular value :', pca.singular_values_)
print('singular vector :\n', pca.components_)
print('explained variance ratio :', pca.explained_variance_ratio_)
re_vector = pca.components_[0]
print(re_vector)
```
